```python
#1)
# Calculate profit percentage
# take the input of cost price and selling price from the user
# determine if there is a profit or loss and calculate the percentage accordingly
# display the result to the user

cost_price = float(input("Enter the cost price: "))
selling_price = float(input("Enter the selling price: "))

if selling_price > cost_price:
    profit = selling_price - cost_price
    profit_percentage = (profit / cost_price) * 100
    print(f"Profit percentage: {profit_percentage:.2f}%")
elif selling_price < cost_price:
    loss = cost_price - selling_price
    loss_percentage = (loss / cost_price) * 100
    print(f"Loss percentage: {loss_percentage:.2f}%")
else:
    print("No profit, no loss.")
```

```python
# 2)
# cricket stats analyzer
# user takes input the run scored by each five player in each match
# each player 1 to 5 ask the user to input the runs scored in each match
# calculate the total runs scored by each player and the average runs per match
# display the total runs and average runs for each player
num_players = 5
num_matches = int(input("Enter the number of matches: "))
players_runs = {}   
for player in range(1, num_players + 1):
    runs = []
    for match in range(1, num_matches + 1):
        run = int(input(f"Enter runs scored by player {player} in match {match}: "))
        runs.append(run)
    players_runs[player] = runs
for player, runs in players_runs.items():
    total_runs = sum(runs)
    average_runs = total_runs / num_matches
    print(f"Player {player}: Total Runs = {total_runs}, Average Runs = {average_runs:.2f}")         
```

```python
# 5)
# WAP to calculate the emplloyee's salary by computing the gross salary tax and net salary based on the following criteria
# take the basic salary as input from the user
# base_salary = 50000
#bonus = 5000
# tax_rate = 10%
#other charges = 2000
# display the gross salary tax and net salary to the user

base_salary = float(input("Enter the basic salary: "))
bonus = 5000
tax_rate = 0.10
other_charges = 2000
gross_salary = base_salary + bonus
tax = gross_salary * tax_rate
net_salary = gross_salary - tax - other_charges

print(f"Gross Salary: {gross_salary:.2f}")
print(f"Tax: {tax:.2f}")
print(f"Net Salary: {net_salary:.2f}")

```

```python
#21)
#WAP for UPSC selection process
upsc_criteria_check = input("Which criteria do you want to check? (Eligibility/Prelims/Mains/Interview): ").strip().lower()

match upsc_criteria_check:
    case "eligibility": 
            age = int(input("Enter your age: "))
            education = input("Do you have a graduation degree? (Yes/No): ").strip().lower()
            nationality = input("Are you a citizen of India? (Yes/No): ").strip().lower()
            if age >= 21 and age <= 32 and education == "yes" and   nationality == "yes":
                print("Congratulations! You are eligible for the UPSC selection process.")
            elif age < 21:
                print("You are not eligible for the UPSC selection process. Age must be at least 21.")
            elif age > 32:
                print("You are not eligible for the UPSC selection process. Age must be 32 or below.")
            elif education != "yes":
                print("You are not eligible for the UPSC selection process. A graduation degree is required.")
            elif nationality != "yes":
                print("You are not eligible for the UPSC selection process. You must be a citizen of India.")
    case "prelims":
            prelims_score = float(input("Enter your prelims score (out of 200): "))
            prelims_cutoff = float(input("Enter the prelims cutoff score (out of 200): "))
            if prelims_score >= prelims_cutoff:
                print("Congratulations! You have qualified for the UPSC mains examination.")
            else:
                print(f"You have not qualified for the UPSC mains examination. You need a score of at least {prelims_cutoff} to qualify.")
    case "mains":
            mains_score = float(input("Enter your mains score (out of 1000): "))
            mains_cutoff = float(input("Enter the mains cutoff score (out of 1000): "))
            if mains_score >= mains_cutoff:
                print("Congratulations! You have qualified for the UPSC interview stage.")
            else:
                print(f"You have not qualified for the UPSC interview stage. You need a score of at least {mains_cutoff} to qualify.")
    case "interview":
            interview_score = float(input("Enter your interview score (out of 300): "))
            interview_cutoff = float(input("Enter the interview cutoff score (out of 300): "))
            if interview_score >= interview_cutoff:
                print("Congratulations! You have successfully cleared the UPSC selection process.")     
            else:
                print(f"You have not cleared the UPSC selection process. You need a score of at least {interview_cutoff} in the interview to clear the process.")
    case _:
            print("Invalid input. Please enter one of the following options: Eligibility, Prelims, Mains, Interview.")  
```

```python
#23)
# WAP to create your own KBC Game

play_game = input("Do you want to play the KBC game? (Yes/No): ").strip().lower()
if play_game == "yes":
    questions = [
        {
            "question": "What is the capital of France?",
            "options": ["A) Berlin", "B) Madrid", "C) Paris", "D) Rome"],
            "answer": "C"
        },
        {
            "question": "Who wrote 'To Kill a Mockingbird'?",
            "options": ["A) Harper Lee", "B) Mark Twain", "C) J.K. Rowling", "D) Ernest Hemingway"],
            "answer": "A"
        },
        {
            "question": "What is the largest planet in our solar system?",
            "options": ["A) Earth", "B) Jupiter", "C) Saturn", "D) Mars"],
            "answer": "B"
        },
        {
            "question": "Who painted the Mona Lisa?",
            "options": ["A) Vincent van Gogh", "B) Pablo Picasso", "C) Leonardo da Vinci", "D) Michelangelo"],
            "answer": "C"
        },
        {
            "question": "What is the chemical symbol for gold?",
            "options": ["A) Au", "B) Ag", "C) Fe", "D) Hg"],
            "answer": "A"
        }
    ]
    
    score = 0
    correct_answers = 0
    skipped_questions = 0
    incorrect_answers = 0
    
    for q in questions:
        print(q["question"])
        for option in q["options"]:
            print(option)
        
        user_input = input("Enter your answer (A/B/C/D), 'skip' to skip, or 'quit' to exit: ").strip().lower()
        
        if user_input == 'quit':
            print("Exiting the game. Goodbye!")
            break
        elif user_input == 'skip':
            skipped_questions += 1
            print("Question skipped.")
        elif user_input.upper() == q["answer"]:
            score += 10
            correct_answers += 1
            print("Correct answer! You earned 10 points.")
        else:
            incorrect_answers += 1
            print("Incorrect answer! No points awarded.")
    
    print("\nGame Over!")
    print(f"Total Score: {score}")
    print(f"Correct Answers: {correct_answers}")
    print(f"Skipped Questions: {skipped_questions}")
    print(f"Incorrect Answers: {incorrect_answers}")
else:
    print("Thank you for your time. Goodbye!")      
    
```
