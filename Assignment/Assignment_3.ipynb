```python
#4) check number it is prime or not
num = int(input("Enter the number:"))
for i in range(2, num):
    if num % i == 0:
        print("Not Prime")
        break
else:
    if num >1:
        print("Prime Number")
    else:
        print("Not Prime")
```

```python
#37 Count Uppercase and lowercase
text = input("Enter string: ")

upper = 0
lower = 0

for ch in text:
    if ch.isupper():
        upper += 1
    elif ch.islower():
        lower += 1

print("Uppercase:", upper)
print("Lowercase:", lower)
```
