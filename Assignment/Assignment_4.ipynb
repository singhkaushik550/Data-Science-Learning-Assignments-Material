```python
# 2.Write a Python program to display the odd and even elements from the list [10, 23, 11, 12, 33, 44, 2, 5, 6]
def display_num(num):
    even=[]
    odd=[]
    for i in num:
        if i % 2 ==0:
            even.append(i)
        else:
            odd.append(i)

    return even, odd

numbers=[10,23, 11, 12, 33, 44, 2, 5, 6]
even, odd = display_num(numbers)
print("Even elements:", even)
print("Odd elements:", odd)
```

```python
# 8. Find the repetitive elements in the list [1,2,3,4,56,1,22,23,33,23, 56].
def repeat_num(nums):
    repeated = []
    for i in nums:
        if nums.count(i) > 1 and i not in repeated:
            repeated.append(i)

    return repeated
numbers = [1, 2, 3, 4, 56, 1, 22, 23, 33, 23, 56]

print("Repeated elements:", repeat_num(numbers))
    
```

```python
# 36.Rearrange a list so that positive and negative numbers alternate.
# List: [1, 2, -3, -4, 5, -6]
# Output:[1, -3, 2, -4, 5, -6]
lst= [1, 2, -3, -4, 5, -6]
positive = []
negative = []
for i in lst:
    if i >=0:
        positive.append(i)
    else:
        negative.append(i)

result = []
for i in range (len(positive)):
    result.append(positive[i])
    if i < len(negative):
        result.append(negative[i])
print(result)
```

```python
# 41.Extract All Digits from a List of Strings
# List:["abc123", "45def", "gh6"]
# Output:["123", "45", "6"]

lst = ["abc123", "45def", "gh6"]
result = []
for i in lst:
    digit = ""
    for j in i:
        if j.isdigit():
            digit += j

    result.append(digit)

print(result)
```

```python
# 42.Find All Anagram Groups in a List of Words.
# List:["eat", "tea", "tan", "ate", "nat", "bat"]
# Output:[["eat", "tea", "ate"], ["tan", "nat"], ["bat"]]
words = ["eat", "tea", "tan", "ate", "nat", "bat"]
anagram_group = {}

for word in words:
    key = "".join(sorted(word))

    if key not in anagram_group:
        anagram_group[key] = []

    anagram_group[key].append(word)

print(list(anagram_group.values()))
```
