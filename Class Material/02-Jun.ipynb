```python
# WAP to check is the string contains vowels or not

word = input("Enter any word or sentences:").lower()
vowel = "a" or "e" or "i" or "o" or "u"

if vowel in word:
    print("Yes it is.")
else:
    print("No it's not.")
```

```python
# WAP to check is valid email '@gmail.com' else invalid email
email = input("Enter your email:").lower()
if "@gmail.com" in email:
    print("Valid email.")
else:    
    print("Invalid email.")
```

```python
# Condidional Expression if,else, elif
# WAP to create a student result system using conditional expression
marks = int(input("Enter your marks:"))
if marks >= 90 and marks <= 100:
    print("A+")
elif marks >= 80 and marks < 90:
    print("A")
elif marks >= 70 and marks < 80:
    print("B")
elif marks >= 60 and marks < 70:    
    print("C")
elif marks >= 50 and marks < 60:
    print("D")
elif marks >= 0 and marks < 50:
    print("F")
else:    
    print("Invalid marks.")
```

```python
# WAP to find the largest number among three numbers using conditional expression
num1 = int(input("Enter first number:"))                
num2 = int(input("Enter second number:"))
num3 = int(input("Enter third number:"))
if num1 > num2 and num1 > num3:
    print("Largest number is:", num1)
elif num2 > num1 and num2 > num3:
    print("Largest number is:", num2)
elif num3 > num1 and num3 > num2:
    print("Largest number is:", num3)
else:
    print("All numbers are equal.")
```

```python
# WAP to find the middle number among three numbers using conditional expression
num1 = int(input("Enter first number:"))
num2 = int(input("Enter second number:"))
num3 = int(input("Enter third number:"))
if (num1 > num2 and num1 < num3) or (num1 < num2 and num1 > num3):
    print("Middle number is:", num1)
elif (num2 > num1 and num2 < num3) or (num2 <  num1 and num2 > num3):
    print("Middle number is:", num2)
elif (num3 > num1 and num3 < num2) or (num3 < num1 and num3 > num2):
    print("Middle number is:", num3)
else:    
    print("All numbers are equal.")
```
