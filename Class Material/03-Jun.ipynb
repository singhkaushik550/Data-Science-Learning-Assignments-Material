```python
# WAP using nested if else for UPSC Exam like  pre, mains and interview.
# min-pre marks = 300
#min-mains marks = 500
#min-interview marks = 700

pre = int(input("Enter pre marks: "))
if pre >= 300:
    print("You have cleared the pre stage.")
    mains = int(input("Enter mains marks: "))
    if mains >= 500:
        print("You have cleared the mains stage.")
        interview = int(input("Enter interview marks: "))              
        if interview >= 700:
            print("You have cleared the UPSC Exam.")
        else:
            print("You did not clear the interview Exam.")
    else:
        print("You did not clear the mains stage.")
else:    
    print("Sorry, you did not clear the pre stage.")   
```

```python
# B.tech Exam Process
# Name Required
# Stream - PCM
# 12th marks - 60% or above

name = input("Enter your name: ").capitalize()
stream = input("Enter your stream: ").upper()
if stream == "PCM":
    marks = float(input("Enter your 12th marks in percentage: "))
    if marks >= 60:
        print(f"Congratulations {name}, you are eligible for B.tech admission.")
    else:
        print(f"Sorry {name}, you are not eligible for B.tech admission due to insufficient marks.")
else:   
     print(f"Sorry {name}, you are not eligible for B.tech admission due to incorrect stream.") 
```
