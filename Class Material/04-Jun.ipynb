## Loops

```python
# for loop
for i in range(1, 11):
    print(i)
```

```python
for i in range(10,0,-1):
    print(i)
```

```python
# find the odd numbers from 1 to 100
for i in range(1, 101):
    if i%2!=0:
        print(i, end=" ")
```

```python
a = [1,2,3,4,5]
for i in a:
    print(i)
```

```python
# sum
total = 0
for i in range(1, 21):  
    total += i 
print(total)    
```

```python
# Table of 2
num = int(input("Enter a number: "))
for i in range(1, 11):
    print(f"{num} x {i} = {num*i}")  
```

```python
n=3
for i in range(1,11):
    prod = n*i
    
    print(f"{n} x {i} = {prod}")
```
