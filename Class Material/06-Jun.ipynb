```python
# create some pattern question using nested for loop 
# create the square
for i in range(3):
    for j in range(3):
        print("*",end=" ")
    print()
```

```python
for i in range(6):
    for j in range(i):
        print("*",end=" ")
    print()
```

```python
for i in range (5,0,-1):
    for j in range(i):
        print("*", end=" ")
    print()
```

```python
n = 4
for i in range(1,n-1):
    print(" " * (n - i), end="")
    
    for j in range(i):
        print("* ", end="")
    
    print()
```
