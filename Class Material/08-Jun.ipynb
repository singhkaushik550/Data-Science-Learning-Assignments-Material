```python
# WAP to take start point and stop point by user and only print all even number
start_point = int(input("Enter the start point number between (1-100):"))
end_point = int(input("Enter the end point number between(1-100):"))

while start_point <= end_point:
    if start_point%2 == 0:
        print(start_point)
    start_point += 1
```

```python
# traversing using while loop
string = "python-programming"
i= 0
while i < len(string):
    print(i,string[i])
    i+=1
```

```python
# reverse the string using while loop
string = "python-programming"
i = len(string)-1
while i >=0:
    print(i,string[i])
    i-=1
```
