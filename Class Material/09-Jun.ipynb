```python
string = "Python-Programming"
i = 0
size = len(string)
count_vowel = 0
consonent = 0
index_sum = 0
while i < size:
    if string[i] in "aeiouAEIOU":
        count_vowel += 1
        
    elif string[i] not in "aeiouAEIOU":
        consonent += 1
    index_sum += i
    i += 1
        
print(f"Total Number of vowels:{count_vowel}")
print(f"Total Number of consonent:{consonent}")
print(f"Sum of index:{index_sum}")

```
