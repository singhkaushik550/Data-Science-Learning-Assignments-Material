```python
# function without return 
# basic function
def addition():
    a=10
    b=20
    c=a+b
    print(c)
addition()
```

```python
# function with single parameter
def even(a=20):
    if a%2 == 0:
        print("Even Number")
    else:
        print("ODD Number")
        
even()
```

```python
# function with double parameter
def multi(a,b):
    mul = a*b
    print(mul)
    
multi(10,20)
```

```python
# function with default parameter
def def_par(a=20,b=5):
    divide = a/b
    print(divide)
    
def_par()
```

```python
# function with return
def add(a,b):
    c=a+b
    return c

res = add(10,20)
print(res)
```

```python
# Create a menue driven program
# addition
# subtraction
# product
# divide

a = int(input("Enter your number 1"))
b = int(input("Enter your number 2"))
choice = input("Enter your choice(+,-,*,/,%)")

def menu():
    if choice == '+':
        return f'Addition of {a}+{b} is:{a+b}'
    elif choice == '-':
        return f'Substraction of {a}-{b} is:{a-b}'
    elif choice == '*':
        return f'Multiplication of {a}*{b} is:{a*b}'
    elif choice == '/':
        return f'Division of {a}/{b} is:{a/b}'
    elif choice == '%':
        return f'Modulus of {a}%{b} is:{a%b}'
    else:
        return 'Invalid choice'
menu()
```
