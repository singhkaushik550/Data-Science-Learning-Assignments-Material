```python
# WAF to extract and store all vowels for string "this-is python"
def extract_vowels(text):
    vowels = ""

    for ch in text:
        if ch == "a" or ch == "e" or ch == "i" or ch == "o" or ch == "u":
            vowels = vowels + ch

    return vowels


string = "this-is python"
result = extract_vowels(string)

print(f"Vowels are: {result}")
```
