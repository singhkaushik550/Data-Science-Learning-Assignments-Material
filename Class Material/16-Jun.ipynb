```python
# Lambda Function
# it is a anonymous function that we will used for breaking the large function into the one line function most of the time 
# it can take any number of argument but only one function

res = lambda a,b:a+b 
result = res(10,20)
print(result)
```

```python
# write a lambda function for add, substract, multiply, divide
trial = lambda x,y: [x+y, x-y, x*y, x/y, x%y]
result = trial(20,10)
print(result)
```

```python
# lambda function with the condition
trial = lambda x: "Even" if x%2 == 0 else "Odd"
result = trial(20)
print(result)
```

```python
# lambda function with the for loop
trial = lambda: [x for x in range(0,11)]
result = trial()
print(result)
```
