```python
# write a function remove last word from string
def remove_last_word(text):
    return text.rsplit(" ", 1)[0]

str1 = "This is python code in vs"

print(remove_last_word(str1))
```

```python
# write a function to remove words if vowel present from string
def remove_vowel_words(text):
    vowels = "aeiouAEIOU"
    return " ".join(
        [word for word in text.split()
         if not any(ch in vowels for ch in word)]
    )

str1 = "This is python code in vs"

print(remove_vowel_words(str1))
```

```python
# write a function to count total whitespace in string
def count_whitespace(text):
    return len([ch for ch in text if ch == " "])

str1 = "This is python code in vs"

print(count_whitespace(str1))
```

```python
# write a function to remove all spaces
str =  "This is python code123 in vs version-35 address noida-11096"

def remove_allspace(a):
    return a.replace(" ", "")

remove_allspace(str)
```

```python
# write a function to print all special symbol.
str = "This is python code@123 in vs version-35 adress noida-11096"

def print_special_symbols(s):
    for ch in s:
        if not ch.isalnum() and ch != " ":
            print(ch)

print_special_symbols(str)
```

```python
# write a function to rverse all words of string word by word
str = "This is python code@123 in vs version-35 add$ress noida-11096"


def reverse_words(s):
    words = s.split()
    return " ".join(word[::-1] for word in words)

print(reverse_words(str))
```
