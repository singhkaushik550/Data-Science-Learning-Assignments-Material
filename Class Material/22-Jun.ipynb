```python
#ex1) write a function to extract all number greater than 55
def check_greater(res):
    lst = []
    for i in res:
        if i >= 55:
            lst.append(i)
    return (tuple(lst))
    
marks_tuple = (50,55,69,34,89)
check_greater(marks_tuple)
```

```python
#ex2 write a function to sum all the indexes
def sum_index(a):
    res = 0
    for i in range(len(a)):
        res+=i
    return res

marks_tuple = (50,55,69,34,89)
sum_index(marks_tuple)
        
```

```python
#ex2 another method
def sum_i(a):
    return sum(range(len(a)))

marks_tuple = (50,55,69,34,89)
sum_i(marks_tuple)
```
