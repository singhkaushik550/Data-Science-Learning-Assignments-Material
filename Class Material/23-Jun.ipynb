```python
emp_name = ["Ram", "rohit", "rohan"]
lst = []
for name in emp_name:
    lst.append(name.upper())
    
print(lst)
```

```python
emp_name = ["Ram", "rohit", "rohan"]
vowel = "aeiou"
count=0
for name in emp_name:
    for ch in name:
        if ch not in vowel:
            print(ch)
            count+=1
print("total vowel:",count)
```
