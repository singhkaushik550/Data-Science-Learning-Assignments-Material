```python
### user max attempt = 6
# each attempt random number generate
# if he attempt that generete random number you can generate it
# you have one fix value 
# if sum equal fix value then winner if low to low and if some difference to close 
import random

fix_value = 30
max_attempt = 6

total = 0

for attempt in range(1, max_attempt + 1):
    num = random.randint(1, 10)
    print(f"Attempt {attempt}: {num}")

    total += num

    if total == fix_value:
        print("Winner")
        break

print("\nFixed Value :", fix_value)
print("Your Sum    :", total)

if total == fix_value:
    print("🏆 Winner")
elif total < fix_value:
    diff = fix_value - total

    if diff <= 3:
        print("😊 Very Close!")
    elif diff <= 7:
        print("👍 Close")
    else:
        print("😔 Too Low")
else:
    diff = total - fix_value

    if diff <= 3:
        print("😊 Very Close!")
    elif diff <= 7:
        print("👍 Close")
    else:
        print("😔 Too High")
```
