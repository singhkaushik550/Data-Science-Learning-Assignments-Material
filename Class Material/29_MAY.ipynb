```python
# Write a program to create the calculater that takes input from a user and gives the output

num1 = float(input("Enter the num1:"))
num2 = float(input("Enter the num2:"))

operation = input("Enter the Operation that you want(+,-,*,/,//,**,%)").strip()

match operation:
    case "+":
        result = num1 + num2
        print(f"{num1} + {num2} = {result}")

    case "-":
        result = num1 - num2
        print(f"{num1} - {num2} = {result}")
    case "*":
        result = num1 * num2
        print(f"{num1} * {num2} = {result}")
    case "/":
        if num2 != 0:
            result = num1/num2
            print(f"{num1} / {num2} = {result}")
        else:
            print("Zero Division Error! Enter number greater than 0.")
    case "//":
        if num2 != 0:
            result = num1//num2
            print(f"{num1} // {num2} = {result}")
        else:
            print("Zero Division Error! Please enter num2 greater than 0.")
    case "**":
        result = num1 ** num2
        print(f"{num1} ** {num2} = {result}")
    case "%":
        if num2 != 0:
            result = num1 % num2
            print(f"{num1} % {num2} = {result}")
        else:
            print(f"Zero Division Error! Please enter num2 greater than 0.")
    case "_":
        print(f"Invalid Operator! Please use these operator (+,-,*,/,//,**,%).")
```
