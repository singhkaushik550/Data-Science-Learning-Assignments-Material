```python
import psycopg2

# Function 1: Create Connection
def create_connection():
    conn = psycopg2.connect(
        host="localhost",
        database="postgres",
        user="postgres",
        password="12345678",
        port="5432"
    )
    return conn


# Function 2: Create Database
def create_database(db_name):
    conn = create_connection()
    conn.autocommit = True

    cur = conn.cursor()

    query = f"CREATE DATABASE {db_name};"

    try:
        cur.execute(query)
        print(f"Database '{db_name}' created successfully.")
    except Exception as e:
        print("Error:", e)

    cur.close()
    conn.close()


# Function 3: Create Table
def create_table():
    conn = create_connection()
    cur = conn.cursor()

    query = """
    CREATE TABLE IF NOT EXISTS students_data (
        s_id SERIAL PRIMARY KEY,
        s_name VARCHAR(100),
        math INT,
        hindi INT,
        cs INT
    );
    """

    cur.execute(query)
    conn.commit()

    print("Table created successfully.")

    cur.close()
    conn.close()


# Function 4: Insert Data
def insert_data(name, math, hindi, cs):
    conn = create_connection()
    cur = conn.cursor()

    query = """
    INSERT INTO students_data (s_name, math, hindi, cs)
    VALUES (%s, %s, %s, %s);
    """

    cur.execute(query, (name, math, hindi, cs))
    conn.commit()

    print("Data inserted successfully.")

    cur.close()
    conn.close()


# Function 5: Read Data
def read_data():
    conn = create_connection()
    cur = conn.cursor()

    cur.execute("SELECT * FROM students_data")

    rows = cur.fetchall()

    for row in rows:
        print(row)

    cur.close()
    conn.close()


# Function 6: Update Data
def update_data(student_id, new_name):
    conn = create_connection()
    cur = conn.cursor()

    query = """
    UPDATE students_data
    SET s_name = %s
    WHERE s_id = %s;
    """

    cur.execute(query, (new_name, student_id))
    conn.commit()

    print("Data updated successfully.")

    cur.close()
    conn.close()


# Function 7: Delete Data
def delete_data(student_id):
    conn = create_connection()
    cur = conn.cursor()

    query = "DELETE FROM students_data WHERE s_id = %s;"

    cur.execute(query, (student_id,))
    conn.commit()

    print("Data deleted successfully.")

    cur.close()
    conn.close()

# Function Calls
# create_database("student_db")
# create_table()
# insert_data("Rohan", 90, 85, 95)
# insert_data("Aman", 88, 80, 92)
# insert_data("Ryan", 80, 90, 79)
# insert_data("Adam", 82, 92, 85)
# insert_data("Jack", 92, 90, 80)

read_data()

# update_data(1, "Rahul")

# delete_data(1)
```

```python
# create a functio and connection with the postgre and create some databases in the given list
import psycopg2

def create_connection():
    conn = psycopg2.connect(
        host="localhost",
        user="postgres",
        password="12345678",
        database="postgres",
        port="5432"
        
    )
    conn.autocommit = True
    return conn


def create_databases(conn, db_list):
    cur = conn.cursor()

    for db in db_list:
        cur.execute(f"CREATE DATABASE {db}")
        print(f"{db} database created")

    cur.close()


target_db_name = ['staff_data', 'students_data', 'employees_data']

conn = create_connection()
create_databases(conn, target_db_name)
conn.close()
```
